```python
hotelalldf = hoteldf.dropna()
hotelalldf['room_info_list'] = hotelalldf['room_info_list'].replace({'\Â':''},regex=True)
hotelalldf['room_info_list'] = hotelalldf['room_info_list'].replace({'\·':','},regex=True)
hotelalldf['occupancy'] = hotelalldf['room_info_list'].astype(str).str.split(" , ").str[0]
hotelalldf['beds'] = hotelalldf['room_info_list'].astype(str).str.split(" , ").str[2]
hotelalldf['bath'] = hotelalldf['room_info_list'].astype(str).str.split(" , ").str[3]
hotelalldf['room_info_list'] = hotelalldf['room_info_list'].astype(str).str.split(" , ").str[1]

def remove_chars(val):
    new_val = re.findall('[0-9.]+',val)
    return "".join(new_val)

hotelalldf.room_info_list = hotelalldf.room_info_list.apply(remove_chars)
hotelalldf.occupancy = hotelalldf.occupancy.apply(remove_chars)
hotelalldf.beds = hotelalldf.occupancy.apply(remove_chars)
hotelalldf.bath = hotelalldf.occupancy.apply(remove_chars)
hotelalldf.rename(columns={'room_info_list': 'rooms'}, inplace = True)
hotelalldf.dropna(axis=0)
print(hotelalldf.shape)
# check nulls
print((hotelalldf.isnull()).sum() )
print((hotelalldf.isna()).sum() )
hotelalldf.sample(5)
```

```python
hotelalldf['prices_list'] = hotelalldf['prices_list'].replace({'\$':''},regex=True)
hotelalldf['prices_list'] = hotelalldf['prices_list'].replace({'\Â':''},regex=True)
hotelalldf.head(5)
```

```python
def clean_price(price):
    price = price[::-1].replace('.', ',').replace(',', '.', 1)[::-1].translate(str.maketrans('', '', "\xa0€,"))
    return price
hotelalldf.prices_list = hotelalldf.prices_list.apply(clean_price)
```

```python
#cleaning hotel star ,grab just numbers with decimal
def remove_chars(val):
    new_val = re.findall('[0-9.]+',val)
    return "".join(new_val)

hotelalldf.stars_list = hotelalldf.stars_list.apply(remove_chars)
hotelalldf.head(5)
```

```python
def remove_special_chars(val):
    new_val = re.findall('[0-9]',val)
    return "".join(new_val)
hotelalldf.number_of_reviews_list = hotelalldf.number_of_reviews_list.apply(remove_special_chars)

hotelalldf.head(5)
```

```python

hotelalldf.stars_list = hotelalldf.stars_list.astype(float)
hotelalldf.prices_list = hotelalldf.prices_list.astype(float)
hotelalldf.number_of_reviews_list = hotelalldf.number_of_reviews_list.astype(int)
hotelalldf.rooms = pd.to_numeric(hotelalldf.rooms,errors='coerce')
hotelalldf.beds = pd.to_numeric(hotelalldf.beds,errors='coerce')
hotelalldf.bath = pd.to_numeric(hotelalldf.bath,errors='coerce')
hotelalldf.occupancy = pd.to_numeric(hotelalldf.occupancy,errors='coerce')
hotelalldf=hotelalldf.dropna(subset=['rooms'])
# check nulls
print((hotelalldf.isnull()).sum() )
print((hotelalldf.isna()).sum() )
```
